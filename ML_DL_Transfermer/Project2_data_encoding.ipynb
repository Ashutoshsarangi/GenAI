```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder

df = pd.DataFrame({"color": ['red', 'green', 'blue', 'red', 'yellow', 'blue']})
```

```python
df.head()
```

```python
# create an instance of OneHotEncoder
encoder = OneHotEncoder()
# fit and transform the data
encoded_data = encoder.fit_transform(df[['color']])
```

```python
encoded_data.toarray()
```

```python
encoded_df = pd.DataFrame(encoded_data.toarray(), columns=encoder.get_feature_names_out(['color']))
```

```python
encoded_df
```

```python
pd.concat([df, encoded_df], axis=1)
```

```python
df = pd.read_csv("./Data/tip.csv")
print(df.head())
```

```python
encoded_sex_df = pd.DataFrame(encoder.fit_transform(df[['sex']]).toarray(), columns=encoder.get_feature_names_out(['sex']))
encoded_smoker_df = pd.DataFrame(encoder.fit_transform(df[['smoker']]).toarray(), columns=encoder.get_feature_names_out(['smoker']))
encoded_day_df = pd.DataFrame(encoder.fit_transform(df[['day']]).toarray(), columns=encoder.get_feature_names_out(['day']))
encoded_time_df = pd.DataFrame(encoder.fit_transform(df[['time']]).toarray(), columns=encoder.get_feature_names_out(['time']))
```
